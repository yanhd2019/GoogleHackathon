```python
def user_input(prompt = "Hi! Ask me an Economics Concept!\n"):
  user_input = input(prompt)
  return user_input
```

```python
def generate_response(text = "Hi"):
  response = palm.generate_text(
    **defaults,
    prompt=text
  )

  result = response.candidates[0]['output']
  return result
```

```python
def segment_response(response):
  rt_val = 0
  if response == "Hi! Ask me an Economics Concept!\n":
    rt_val = -1 #User says hi
    text = response
    code = ""
  elif response == "This is not an Economic concept. Ask me anything about Economics instead!\n":
    rt_val = -2 #user ask non econ stuff
    text = response
    code = ""
  else
  elif response == "":
    rt_val = -3 #API gives up
    text = response
    code = ""
  else:
    if "### Explanation" in response:
      if "### Figure" in response:
        rt_val = 0 #Normal case
        code, text = response.split("### Explanation\n")
        code = code.split()
      else:
        rt_val = 1 #Case no figure

    else:
      if "### Figure" in response:
        rt_val = 2 #Case no Explanation


      else:
        rt_val = 3 #Case WHAT THERE IS NOTHING YET SOMETHING
        text = response
        code = ""

  return rt_val, text, code

```
